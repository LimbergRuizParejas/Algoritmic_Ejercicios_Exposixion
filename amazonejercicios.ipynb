```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
```

```python
data = pd.read_csv('/kaggle/input/data-set/amazon.csv')
data.head()
```

```python
# Seleccionar características y etiqueta
features = ['discounted_price', 'rating']
X = data[features]
y = data['category']
```

```python
# Codificar la columna de etiqueta
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Escalar las características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Función para evaluar K-NN con diferentes valores de 'k'
def knn_with_k(k):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    
```

```python
accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
    
    print(f"Results for k={k}:")
    print("Accuracy:", accuracy)
    print("Classification Report:\n", report)
```

```python
# Crear un DataFrame para las visualizaciones de prueba y predicción
    visualization_df = pd.DataFrame(X_test, columns=['discounted_price', 'rating'])
    visualization_df['true_label'] = label_encoder.inverse_transform(y_test)
    visualization_df['predicted_label'] = label_encoder.inverse_transform(y_pred)
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='discounted_price', y='rating', hue='true_label', style='predicted_label', data=visualization_df)
    plt.title(f"K-NN with k={k}")
    plt.xlabel('Precio Descontado')
    plt.ylabel('Rating')
    plt.legend(title='Categoría')
    plt.show()
```

```python
# Ejemplo de uso con k=3, k=5, y k=7
for k in [3, 5, 7]:
    knn_with_k(k)
```
